```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
from scipy.stats import beta as beta_dist

np.random.seed(0)

# ================================================================
# SIGNAL GENERATION
# ================================================================
sr         = 44100
duration   = 3.0
frequency  = 440.0

t = np.linspace(0, duration, int(sr * duration), endpoint=False)
clean_signal = 0.5 * np.sin(2 * np.pi * frequency * t)

# ================================================================
# CORE FUNCTIONS (YIN)
# ================================================================
def difference_function(frame, max_tau):
    W  = len(frame) // 2
    df = np.zeros(max_tau)
    for tau in range(1, max_tau):
        diff    = frame[:W] - frame[tau:tau + W]
        df[tau] = np.sum(diff ** 2)
    return df

def cmnd_func(df):
    out         = np.ones(len(df))
    running_sum = 0.0
    for tau in range(1, len(df)):
        running_sum += df[tau]
        out[tau] = 1.0 if running_sum == 0 else df[tau] * tau / running_sum
    return out

def parabolic_interp(c, tau):
    if tau <= 1 or tau >= len(c) - 1:
        return float(tau)
    s0, s1, s2 = c[tau-1], c[tau], c[tau+1]
    denom      = 2.0 * (2.0 * s1 - s2 - s0)
    return float(tau) if abs(denom) < 1e-10 else tau + (s2 - s0) / denom

def pick_tau(c, threshold):
    for tau in range(2, len(c) - 1):
        if c[tau] < threshold and c[tau] <= c[tau-1] and c[tau] <= c[tau+1]:
            return tau
    return int(np.argmin(c[2:]) + 2)

def run_yin(y, sr, frame_size=2048, hop_size=256, threshold=0.1):
    y = np.pad(y, (0, frame_size), mode='constant')
    frames = librosa.util.frame(y, frame_length=frame_size, hop_length=hop_size)

    f0 = np.zeros(frames.shape[1])
    max_tau = frame_size // 2

    for i in range(frames.shape[1]):
        df_ = difference_function(frames[:, i], max_tau)
        c   = cmnd_func(df_)
        tau = pick_tau(c, threshold)
        tau_r = parabolic_interp(c, tau)
        f0[i] = sr / tau_r if tau_r > 0 else 0.0

    return f0

# ================================================================
# ACCURACY
# ================================================================
def measure_accuracy(f0, true_freq, tol=20):
    valid = f0 > 0
    if np.sum(valid) == 0:
        return 0.0
    return 100.0 * np.sum(np.abs(f0[valid] - true_freq) < tol) / np.sum(valid)

# ================================================================
# PYIN STAGE 1
# ================================================================
def pyin_stage1_frame(frame, sr, frame_size=2048, alpha=1, beta_p=18):
    max_tau = frame_size // 2
    df_ = difference_function(frame, max_tau)
    c   = cmnd_func(df_)

    thresholds = np.arange(0.01, 1.01, 0.01)
    weights    = beta_dist.pdf(thresholds, alpha, beta_p)
    weights   /= weights.sum()

    tau_probs = {}

    for s, w in zip(thresholds, weights):
        tau = pick_tau(c, s)
        tau_probs[tau] = tau_probs.get(tau, 0.0) + w

    candidates = []
    for tau, prob in tau_probs.items():
        tau_r = parabolic_interp(c, tau)
        if tau_r <= 0: continue
        freq = sr / tau_r
        if 50.0 <= freq <= 2000.0:
            candidates.append((freq, prob))

    min_c = float(c[2:].min())
    voiced_prob = float(np.clip(0.95 - 1.7 * min_c, 0.05, 0.95))
    unvoiced_prob = 1.0 - voiced_prob

    total_w = sum(p for _, p in candidates)
    if total_w > 0:
        candidates = [(f, (p/total_w)*voiced_prob) for f, p in candidates]

    return candidates, voiced_prob, unvoiced_prob

def run_pyin_stage1(y, sr, frame_size=2048, hop_size=256):
    y = np.pad(y, (0, frame_size), mode='constant')
    frames = librosa.util.frame(y, frame_length=frame_size, hop_length=hop_size)

    num_frames = frames.shape[1]
    all_cands = []
    all_p_unvoiced = np.zeros(num_frames)

    for i in range(num_frames):
        cands, vp, uvp = pyin_stage1_frame(frames[:, i], sr, frame_size)
        all_cands.append(cands)
        all_p_unvoiced[i] = uvp

    return all_cands, all_p_unvoiced

# ================================================================
# HMM + VITERBI
# ================================================================
NUM_BINS = 240
F_MIN, F_MAX = 50.0, 1000.0
NUM_STATES = 2 * NUM_BINS

pitch_bins = F_MIN * (F_MAX/F_MIN) ** (np.arange(NUM_BINS)/NUM_BINS)

def freq_to_bin(freq):
    if freq < F_MIN or freq > F_MAX:
        return -1
    return int(np.clip(
        round(NUM_BINS * np.log2(freq/F_MIN) / np.log2(F_MAX/F_MIN)),
        0, NUM_BINS-1))

def build_A(num_bins):
    N = 2 * num_bins
    A = np.zeros((N, N))

    for i in range(num_bins):
        A[i, i] = 0.9
        if i+1 < num_bins:
            A[i, i+1] = 0.05
        if i-1 >= 0:
            A[i, i-1] = 0.05
        A[i, num_bins+i] = 0.01

    for i in range(num_bins):
        A[num_bins+i, num_bins+i] = 0.9
        A[num_bins+i, i] = 0.1

    return A

def build_obs(all_cands, all_p_unvoiced, num_bins):
    num_frames = len(all_cands)
    obs = np.zeros((num_frames, 2*num_bins))

    for t, (cands, p_uv) in enumerate(zip(all_cands, all_p_unvoiced)):
        for freq, prob in cands:
            b = freq_to_bin(freq)
            if b >= 0:
                obs[t, b] += prob
        obs[t, num_bins:] = p_uv / num_bins
        obs[t] += 1e-10
        obs[t] /= obs[t].sum()

    return obs

def viterbi(obs, A, pi):
    T, N = obs.shape
    log_A = np.log(A + 1e-300)
    log_obs = np.log(obs + 1e-300)
    log_pi = np.log(pi + 1e-300)

    dp = np.full((T, N), -np.inf)
    psi = np.zeros((T, N), dtype=np.int32)

    dp[0] = log_pi + log_obs[0]

    for t in range(1, T):
        scores = dp[t-1][:, np.newaxis] + log_A
        psi[t] = np.argmax(scores, axis=0)
        dp[t] = np.max(scores, axis=0) + log_obs[t]

    path = np.zeros(T, dtype=np.int32)
    path[-1] = np.argmax(dp[-1])

    for t in range(T-2, -1, -1):
        path[t] = psi[t+1, path[t+1]]

    return path

def path_to_f0(path):
    f0 = np.zeros(len(path))
    for i, s in enumerate(path):
        if s < NUM_BINS:
            f0[i] = pitch_bins[s]
    return f0

def run_full_pyin(y, sr):
    all_cands, all_p_unvoiced = run_pyin_stage1(y, sr)
    A = build_A(NUM_BINS)
    obs = build_obs(all_cands, all_p_unvoiced, NUM_BINS)

    pi = np.ones(NUM_STATES) / NUM_STATES

    path = viterbi(obs, A, pi)
    return path_to_f0(path)

# ================================================================
# TESTING
# ================================================================
noise_levels = [0.0, 0.05, 0.1, 0.15, 0.2]

yin_accs = []
pyin_accs = []

print(f"{'Noise':>6} {'YIN%':>8} {'PYIN%':>8}")

for noise in noise_levels:
    noisy = clean_signal + noise * np.random.randn(len(clean_signal))

    yin_f0 = run_yin(noisy, sr)
    pyin_f0 = run_full_pyin(noisy, sr)

    yin_acc = measure_accuracy(yin_f0, frequency)
    pyin_acc = measure_accuracy(pyin_f0, frequency)

    yin_accs.append(yin_acc)
    pyin_accs.append(pyin_acc)

    print(f"{noise:>6.2f} {yin_acc:>8.1f} {pyin_acc:>8.1f}")

# ================================================================
# PLOT
# ================================================================
plt.plot(noise_levels, yin_accs, 'o-', label='YIN')
plt.plot(noise_levels, pyin_accs, 's-', label='PYIN')
plt.xlabel('Noise Level')
plt.ylabel('Accuracy (%)')
plt.title('YIN vs PYIN')
plt.legend()
plt.grid()
plt.show()

print("\n✅ Final code working successfully!")
```
